```python
## SCIPY

fs2, audio2 = wav.read(filename)
print("Shape and type of the Audio data x:", audio2.shape, type(audio2), "SR: ", fs2)

tot_secs_len = len(audio2)/fs2
tot_mins_len = tot_secs_len/60

print("Seconds:", tot_secs_len, "Mins:", tot_mins_len)

audio2
```
